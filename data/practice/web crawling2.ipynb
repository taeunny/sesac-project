#### 문제 2: 수집한 `top_movies` 리스트 안에 있는 딕셔너리에서 `code`, `title`, `year`, `ratings_avg`, `nations`, `genres`만 남기고 나머지는 필요 없으니 없애버리세요.

```python
new_top_movies = []
for tm in top_movies:
    new_top_movies.extend(tm)

```

```python
len(new_top_movies)
```

```python
top_movies = new_top_movies
```

```python
cleaned_top_movies = []
for movie in top_movies:
    code = movie['code']
    title = movie['title']
    year = movie['year']
    ratings_avg = movie['ratings_avg']
    nations = [m['name'] for m in movie['nations']]
    genres = movie['genres']
    movie_dict = {
        "code" : code,
        "title" : title,
        "year" : year,
        "ratings_avg" : ratings_avg,
        "nations" : nations,
        "genres" : genres
    }
    cleaned_top_movies.append(movie_dict)
    
```

#### 문제3: `cleaned_top_movies`를 데이터 프레임으로 만들고, `watcha_top_movies.csv` 파일로 저장하세요.

```python
top_movies_df = pd.DataFrame(cleaned_top_movies)
top_movies_df
```

```python
# 중복된 요소 제거
# inplace가 아니기 때문에 다시 받아줘야 한다
top_movies_df = top_movies_df.drop_duplicates(['code'])
top_movies_df  = top_movies_df.reset_index().iloc[:, 1:]     # 행은 다 가져오고 열은 1번자리부터 가져옴
top_movies_df.to_csv("watcha_top_movies.csv", index=False)
```

```python
top_movies_df['nations']= top_movies_df['nations'].apply(lambda x:",".join(x))
```
